```python
import email
import re, json, html2text
from datetime import datetime
from typing import Tuple, Optional

from dateutil import parser

def parse_date(date_str: str) -> Tuple[datetime, datetime]:
    r = re.findall(f'\d+', date_str)
    l = date_str.split()[2]
    month = ['', 'января', "февраля", "марта", "апреля", "мая", "июня", "июля", "августа", "сентября", "октября",
             "ноября", "декабря"]
    begin = f"{r[0]}/{month.index(l)}/{r[1]} {r[2]}:{r[3]}"
    end = f"{r[0]}/{month.index(l)}/{r[1]} {r[4]}:{r[5]}"
    begin_date_time_obj = parser.parse(begin)
    end_date_time_obj = parser.parse(end)
    return json.dumps(begin_date_time_obj, cls=DateTimeEncoder), json.dumps(end_date_time_obj, cls=DateTimeEncoder)
```

```python
from dateutil import parser
def extract_date(l: list):
    dates = []
    months = ['января', "февраля", "марта", "апреля", "мая", "июня", "июля", "августа", "сентября", "октября", 
         "ноября", "декабря"]
    for i, item in enumerate(l):
        for index,m in enumerate(months):
            if m in item:
                try:
                    d = re.match(r'\d+', item).group()
                    m = index+1
                    y = re.search(r'\d{4}', item).group()
                    regex = r'\d{1,2}:\d{2}'
                    begin_str = re.findall(regex, item)[0]
                    end_str = re.findall(regex, item)[0]
                    begin_obj = [parser.parse(f"{d}/{m}/{y} {begin_str}")]
                    end_to = [parser.parse(f"{d}/{m}/{y} {begin_str}")]
                    from_to_dates = [begin_obj, end_to]
                    dates.append(from_to_dates)
                except AttributeError:
                    pass
    return dates
```
